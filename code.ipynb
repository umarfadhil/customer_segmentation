```python
import pandas as pd

df = pd.read_json('dataset.json')
```

```python
import json

x = []
y = []

for i in df['Data']:
    json_acceptable_string = i.replace("'", "\"")
    x.append(json_acceptable_string)

for i in x:
    dd = json.loads(i)
    y.append(dd)
    # dd = {k:[v] for k,v in dd.items()}

data = pd.DataFrame.from_dict(y)
```

```python
df = pd.concat([df, data.reindex(df.index)], axis = 1).drop(columns = 'Data')
```
